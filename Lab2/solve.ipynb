```python
def compress_string(s):
    """
    Сжимает строку, заменяя последовательности одинаковых символов
    на количество и сам символ.
    """
    if not s:
        return ""

    compressed_s = []
    count = 1
    
    for i in range(len(s) - 1):
        if s[i] == s[i+1]:
            count += 1
        else:
            if count > 1:
                compressed_s.append(str(count))
            compressed_s.append(s[i])
            count = 1
    
    if count > 1:
        compressed_s.append(str(count))
    compressed_s.append(s[-1])
    
    return "".join(compressed_s)

input_data = 'AAACBBBBBDDADDD'

output_data = compress_string(input_data)
print(f"Входные данные: {input_data}")
print(f"Выходные данные: {output_data}")

# --- Для самостоятельного ввода ---
# user_input = input("Введите строку для сжатия: ")
# print(f"Результат: {compress_string(user_input)}")
```

```python
def is_password_valid(password):
    """
    Проверяет пароль на соответствие правилам надежности.
    """
    SPECIAL_CHARS = set(['!', '@', '#', '$', '%', '&', '*', '+'])
    
    if len(password) < 12:
        return False

    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for char in password:
        if 'a' <= char <= 'z':
            has_lower = True
        elif 'A' <= char <= 'Z':
            has_upper = True
        elif '0' <= char <= '9':
            has_digit = True
        elif char in SPECIAL_CHARS:
            has_special = True
            
    return has_upper and has_lower and has_digit and has_special

n = 2
passwords = [
    "123456789aA!",
    "123IsAVeryDifficultPassword"
]

print("Входные данные:")
print(n)
for p in passwords:
    print(p)

print("\nВыходные данные:")
for password in passwords:
    if is_password_valid(password):
        print("Valid")
    else:
        print("Invalid")

# --- Для самостоятельного ввода ---
# try:
#     num_passwords = int(input("Введите количество паролей: "))
#     for _ in range(num_passwords):
#         pwd = input("Введите пароль для проверки: ")
#         if is_password_valid(pwd):
#             print("Valid")
#         else:
#             print("Invalid")
# except ValueError:
#     print("Некорректное количество паролей.")
```

```python
import math

def get_median(sorted_data):
    """
    Вычисляет медиану (Q2) для отсортированного списка.
    """
    n = len(sorted_data)
    if n == 0:
        return 0
    mid = n // 2
    if n % 2 == 1:
        return sorted_data[mid]
    else:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2

def count_outliers(data):
    """
    Подсчитывает количество выбросов в наборе данных.
    """
    data.sort()
    n = len(data)
    
    q2 = get_median(data)
    
    mid_index = n // 2
    if n % 2 == 1:
        lower_half = data[:mid_index]
        upper_half = data[mid_index+1:]
    else:
        lower_half = data[:mid_index]
        upper_half = data[mid_index:]

    q1 = get_median(lower_half)
    q3 = get_median(upper_half)
    
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    outliers_count = 0
    for value in data:
        if value < lower_bound or value > upper_bound:
            outliers_count += 1
            
    return outliers_count


n = 9
attendance_data = [25, 24, 19, 99, 18, 17, 0, 21, 26]

print("Входные данные:")
print(n)
for val in sorted(attendance_data):
    print(val)

result = count_outliers(attendance_data)
print(f"\nВыходные данные:\n{result}")

# --- Для самостоятельного ввода ---
# try:
#     count = int(input("Введите количество записей: "))
#     data_list = []
#     for _ in range(count):
#         data_list.append(int(input("Введите число: ")))
#     print(f"Количество некорректных записей: {count_outliers(data_list)}")
# except ValueError:
#     print("Ошибка ввода. Вводите только целые числа.")
```

```python
def calculate_transactions(participants, purchases):
    """
    Рассчитывает минимальное количество транзакций для выравнивания затрат.
    """
    balances = {p: 0.0 for p in participants}

    total_spent = 0
    for payer, amount in purchases:
        balances[payer] += amount
        total_spent += amount
    
    if not participants:
        return 0, []
    avg_share = total_spent / len(participants)

    debtors = []
    creditors = []
    for person, spent in balances.items():
        balance = spent - avg_share
        if balance > 0.01:
            creditors.append({'name': person, 'amount': balance})
        elif balance < -0.01:
            debtors.append({'name': person, 'amount': -balance})

    transactions = []
    while debtors and creditors:
        debtor = debtors[0]
        creditor = creditors[0]

        transfer_amount = min(debtor['amount'], creditor['amount'])

        transactions.append(
            f"{debtor['name']} {creditor['name']} {transfer_amount:.2f}"
        )

        debtor['amount'] -= transfer_amount
        creditor['amount'] -= transfer_amount

        if debtor['amount'] < 0.01:
            debtors.pop(0)
        if creditor['amount'] < 0.01:
            creditors.pop(0)

    return len(transactions), transactions



names = "Ivan Aleksej Igor".split()
n = 3
purchase_list = [
    ("Ivan", 500),
    ("Aleksej", 100),
    ("Ivan", 200)
]

m, transfers = calculate_transactions(names, purchase_list)

print("Выходные данные:")
print(m)
for t in transfers:
    print(t)
```

```python
from collections import Counter

def analyze_pizza_orders(orders):
    """
    Анализирует список заказов и выводит статистику.
    """
    if not orders:
        print("Нет данных для анализа.")
        return

    pizza_names = [order['name'] for order in orders]
    pizza_counts = Counter(pizza_names)
    
    print("--- Популярность пицц (по убыванию) ---")
    for name, count in pizza_counts.most_common():
        print(f"{name}: {count} раз(а)")
        
    daily_sales = {}
    for order in orders:
        date = order['date']
        price = order['price']
        daily_sales[date] = daily_sales.get(date, 0) + price
    
    print("\n--- Продажи по дням (хронологически) ---")
    for date, total in sorted(daily_sales.items()):
        print(f"{date}: {total:.2f}")

    most_expensive_order = max(orders, key=lambda x: x['price'])
    print("\n--- Самый дорогой заказ ---")
    print(f"Дата: {most_expensive_order['date']}, "
          f"Пицца: {most_expensive_order['name']}, "
          f"Стоимость: {most_expensive_order['price']:.2f}")
    
    total_cost = sum(order['price'] for order in orders)
    average_cost = total_cost / len(orders)
    print("\n--- Средняя стоимость заказа ---")
    print(f"{average_cost:.2f}")



input_orders_raw = [
    "2025-09-10 Маргарита 350",
    "2025-09-10 Пепперони 450.50",
    "2025-09-11 Маргарита 360",
    "2025-09-11 Четыре сыра 550",
    "2025-09-10 Пепперони 460",
]

parsed_orders = []
print("--- Обработка входных данных ---")
for line in input_orders_raw:
    parts = line.split()
    date = parts[0]
    price = float(parts[-1])
    name = " ".join(parts[1:-1])
    
    parsed_orders.append({'date': date, 'name': name, 'price': price})
    print(f"Распознан заказ: Дата='{date}', Пицца='{name}', Цена={price}")

print("\n--- Результаты анализа ---\n")
analyze_pizza_orders(parsed_orders)

# --- Для самостоятельного ввода с защитой от ошибок ---
# print("\nВведите данные о заказах (ГГГГ-ММ-ДД Название Цена).")
# print("Для завершения введите пустую строку.")
# user_orders = []
# while True:
#     line = input()
#     if not line:
#         break
#     try:
#         parts = line.split()
#         if len(parts) < 3:
#             print("Неверный формат. Нужно как минимум 3 элемента: дата, название, цена.")
#             continue
#         
#         date = parts[0]
#         price = float(parts[-1])
#         name = " ".join(parts[1:-1])
#         user_orders.append({'date': date, 'name': name, 'price': price})
#     except (ValueError, IndexError):
#         print("Неверный формат. Попробуйте еще раз.")
# 
# analyze_pizza_orders(user_orders)
```

```python
import requests
from collections import Counter

def find_most_active_author(url):
    """
    Находит самого активного автора в почтовом логе mbox.
    """
    try:
        response = requests.get(url)
        response.raise_for_status()
        mbox_text = response.text
    except requests.exceptions.RequestException as e:
        print(f"Ошибка при загрузке файла: {e}")
        return

    author_emails = []

    for line in mbox_text.split('\n'):
        if line.startswith('From '):
            parts = line.split()
            if len(parts) > 1:
                author_emails.append(parts[1])

    if not author_emails:
        print("Не найдено ни одного автора.")
        return
        
    email_counts = Counter(author_emails)
    
    most_common = email_counts.most_common(1)[0]
    
    author, count = most_common
    print(f"Больше всех писем ({count}) отправил автор с адресом: {author}")



file_url = 'https://www.py4e.com/code3/mbox.txt'
find_most_active_author(file_url)
```
